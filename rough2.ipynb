```python
#Using np.select() for multiple conditions
import numpy as np

conditions = [
    (df['Rating'] >= 8),
    (df['Rating'] >= 5) & (df['Rating'] < 8),
    (df['Rating'] >= 3) & (df['Rating'] < 5),
    (df['Rating'] >= 0) & (df['Rating'] < 3)

]
choices = ['a+', 'a', 'b+','b']
df['grade_multi'] = np.select(conditions, choices, default='Na')
```
